# Modern Automated LBO Demo 🚀
This notebook demonstrates the automated LBO modeling library using AAPL, MSFT, and TSLA.

```python
import sys
import os
sys.path.append(os.path.abspath("../src"))
from automated_lbo import LBOModel, ExcelWriter

tickers = ["AAPL", "MSFT", "TSLA"]
for ticker in tickers:
    print(f"Running LBO for {ticker}...")
    model = LBOModel(ticker)
    returns = model.run(equity_pct=0.3, entry_multiple=12.0, exit_multiple=12.0)
    print(f"{ticker} IRR: {returns['IRR']*100:.2f}%")
    
    writer = ExcelWriter(model)
    writer.save(f"{ticker}_LBO_Model.xlsx")
```
